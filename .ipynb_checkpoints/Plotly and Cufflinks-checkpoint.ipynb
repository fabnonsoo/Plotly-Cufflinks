## Plotly and Cufflinks
Plotly is a library that allows us to create interactive plots that you can use in dashboards or websites (you can save them as html files or static images).

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# Knowing the plotly version

from plotly import __version__
print(__version__)
```

```python
import cufflinks as cf
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
# Connecting JavaScript to the notebook, since plotly essentially connects Pandas and Python to an interactive JavaScript library...
# Also allows our notebook to access those visualizations...

init_notebook_mode(connected=True)
```

```python
# Allows us to use cufflinks offline..

cf.go_offline()
```

```python
# Creating a Data with 100 rows and 4 columns....

df1 = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df1.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df2.head()
```

```python
# Ex 1:- Creating our first plot...

df1.plot()
```

```python
# To see our plot converted to a plotly (line plot) interactive image: we use iplot()

df1.iplot()
```

## Using Cufflinks & Plotly
    * scatter

### Scatter Plot

```python
# Plotting this line just gives us a plot w/connected lines. To get our scatter plot, we include the "mode" argument...

# Ex 2i:-

df1.iplot(kind='scatter',x='A',y='B')
```

```python
# Ex 2ii:-

df1.iplot(kind='scatter',x='A',y='B', mode = "markers")
```

```python
# Ex 2iii:- Paasing the size args to make the data points bigger/smaller

df1.iplot(kind='scatter',x='A',y='B', mode='markers', size=10)
```

### Bar Plot

```python
# We're using df2 because it contains categoriaclly columns and suits the plotting of a Bar Plot...
# In a categorical plot, x-axis always takes the categories while y-axis takes the numerical data...

# Ex 3i:-

df2.iplot(kind='bar',x='Category',y='Values')
```

```python
# Since df1 isn't a categorically data, inorder to get a bar plot for such DataFrame(df1), we use an aggregrate function....
# You see that the chart plots bars for each data point which isn't nice. Hence, we need to pass in aggregrate functions like sum, count etc...
# See Ex 3iii

# Ex 3ii:- 

df1.iplot(kind='bar')
```

```python
# Ex 3iii:- Counting the instances for each column...

df1.count().iplot(kind='bar')
```

```python
# Ex 3iv:- Taking the sum of each column.....

df1.sum().iplot(kind='bar')
```

### Box Plot

```python
# Ex 4:-

df1.iplot(kind='box')
```

### 3D-Surface Plot

```python
# Ex 5:-

df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[800,700,600,500,400]})
df3
```

```python
# Creating a 3D-Surface Plot...

df3.iplot(kind="surface")
```

```python
# Ex 5ii:-

df4 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[9,8,7,6,5]})
df4
```

```python
# Creating a 3D-Surface Plot...

df4.iplot(kind="surface")
```

```python
# Ex 5iii:- Changing the color scheme of the df4 surface plot...

df3.iplot(kind="surface", colorscale="rdylbu")
```

### Histogram

```python
# Ex 6i:- Passing in one column (A) from the DataFrame..

df1['A'].iplot(kind='hist')
```

```python
# Ex 6ii:- Specifying the number of bins....

df1['A'].iplot(kind='hist', bins=55)
```

```python
# Ex 6iii:- Passing the entire DataFrame(df1)...

df1.iplot(kind='hist')
```

### Spread

```python
# Ex 7:- The Spread Plot are used a lot for stock data. Especially when you want to see the spread betweeen two stocks..
    
df1[['A','B']].iplot(kind='spread')
```

### Bubble Plot
Similar to scatter plot. You can see this kind of plots for data representing World GDP in comparison to population or happiness factor.

```python
# Ex 8:- Here, the size="C" is based on the data points of column C

df1.iplot(kind='bubble', x ="A", y="B", size="C")
```

### scatter_matrix()
Similar to seaborn's pairplot: sns.pairplot()

```python
# In plotting a scatter matrix, you must make sure all your columns in the DataFrame(df1) are numerical for the scatter matrix to plot..
# It isn't good to use a Scatter matrix when you have lots of data points beacuse it'll take a lot of time to load and result in crashing your python kernel..

# Ex 9:-

df1.scatter_matrix()
```
